```python
import random
import pandas as pd

# Create a list of random names for stylists
first_names = ["Hyejin", "Minji", "Seojun", "Jihun", "Yeji", "Taeyang", "Dasom", "Gyuri", "Eunwoo", "Soyeon"]
last_names = ["Kim", "Lee", "Park", "Choi", "Jung", "Kang", "Yoon", "Im", "Moon", "Shin"]
names = [f"{random.choice(first_names)} {random.choice(last_names)}" for _ in range(100)]

# Create a list of random salon names
salon_names = [f"Salon {i+1}" for i in range(50)]
salons = [random.choice(salon_names) for _ in range(100)]

# Randomize followers count, post texts, and likes
follower_counts = [random.randint(5000, 50000) for _ in range(100)]
post_texts = [
    "Gorgeous waves 🌊", "Sleek and shiny hair 💇‍♀️", "Bold new look 🔥", 
    "Amazing hair transformation!", "How to maintain curls 🌟", 
    "Classic bob, reimagined.", "Hair treatment tips!", "Salon day fun ✨"
]
popular_posts = [random.choice(post_texts) for _ in range(100)]
likes = [random.randint(500, 10000) for _ in range(100)]

# Create a DataFrame for Instagram Accounts Data
df_stylists = pd.DataFrame({
    "Name": names,
    "Salon Name": salons,
    "Followers": follower_counts,
    "Most Popular Post Text": popular_posts,
    "Likes": likes
})

# Create a Following Relationship Data
following_relationships = []
for follower in names:
    num_following = random.randint(1, 10)  # Each stylist follows 1 to 10 accounts
    followed = random.sample(names, num_following)
    for followed_account in followed:
        if follower != followed_account:  # Avoid self-follow
            following_relationships.append((follower, followed_account))

# Create a DataFrame for Following Relationship
df_relationships = pd.DataFrame(following_relationships, columns=["Follower", "Followed"])

# Save both DataFrames for reference
df_stylists.to_csv("stylists_data.csv", index=False)
df_relationships.to_csv("following_relationships.csv", index=False)

# Print to verify
df_stylists.to_csv("stylists_data.csv")
df_relationships.to_csv("following_relationships.csv")
```
